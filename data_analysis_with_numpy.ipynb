## Dataset Description

Attribute Information -

The dataset contains 517 instances and 13 attributes. The attribute information is as follows:-

* X - x-axis spatial coordinate within the Montesinho park map: 1 to 9
* Y - y-axis spatial coordinate within the Montesinho park map: 2 to 9
* month - month of the year: 'jan' to 'dec'
* day - day of the week: 'mon' to 'sun'
* FFMC - FFMC index from the FWI system: 18.7 to 96.20
* DMC - DMC index from the FWI system: 1.1 to 291.3
* DC - DC index from the FWI system: 7.9 to 860.6
* ISI - ISI index from the FWI system: 0.0 to 56.10
* temp - temperature in Celsius degrees: 2.2 to 33.30
* RH - relative humidity in %: 15.0 to 100
* wind - wind speed in km/h: 0.40 to 9.40
* rain - outside rain in mm/m2 : 0.0 to 6.4
* area - the burned area of the forest (in ha): 0.00 to 1090.84

# NumPy ndarray object

numpy.array(object, dtype=None, copy=True, order=None, subok=False, ndmin=0)

The above constructor takes the following parameters:
* object - Any object exposing the array interface method returns an array, or any (nested) sequence.
* dtype - Desired data type of array, optional.
* copy - (Optional) - By default (true), the object is copied.
* order - C (row major) or F (column major) or A (any) (default).
* subok - By default, returned array forced to be a base class array. If true, sub-classes passed through.
* ndmin - Specifies minimum dimensions of resultant array.

```python
# one-dimensional array
a = np.array([1,2,3])
print(a)
```

```python
n1 = np.array(fires[:1], dtype=np.float32)
print(n1)
```

```python
# two-dimensional array
b = np.array([
    [1,2],
    [3,4]
])

print(b)
```

```python
n2 = np.array(fires[:2], dtype=int)

print(n2)
```

```python
# there-dimensional array
c = np.array([
    [1,2],
    [3,4],
    [5,6]
])

print(c)
```

```python
n3 = np.array(fires[:3], dtype=np.int64)

print(n3)
```

```python
# minimum dimensions

d = np.array([1,2,3,4,5], ndmin=2)
print(d)
```

```python
# dtype parameter

e = np.array([1,2,3,4,5], dtype=complex)

print(e)
```

## NumPy data types

```python
dt = np.dtype(np.int32)

print(dt)
```

```python
# create structured data type
dt = np.dtype([('age', np.int8)])
print(dt)
```

```python
# now apply it to ndarray object
dt = np.dtype([('age', np.int8)])

arr1 = np.array([(10,), (20,), (30,)], dtype=dt)
print(arr1)
```

```python
# file name can be used to access content of age column
dt = np.dtype([('age', np.int8)])

arr2 = np.array([(10,), (20,), (30,)], dtype=dt)
print(arr2['age'])
```

```python
student = np.dtype([('name', 'S20'), ('age', 'i1'), ('marks', 'f4')])

print(student)
```

```python
student = np.dtype([('name', 'S20'), ('age', 'i1'), ('marks', 'f4')])
student_data = np.array([('Krasi', 23, 50), ('Vlado', 24, 55)], dtype=student)
print(student_data)
```

## NumPy array creation

A new ndarray object can be constructed by any of the following array creation routines. There are several ways to create arrays. These are listed below:
* array function
* empty function
* zeros function
* ones function
* arange function
* linspace function

### array function

```python
x7 = np.array([1,2,3])
x7
```

```python
x7.dtype
```

```python
list1 = [1,2,3,4]

a1 = np.array(list1)
print(a1.dtype)
```

```python
dict1 = {
    "a": 1,
    "b": 2,
    "c": 3,
}

a2 = np.array(dict1, dtype=np.object_)
print(a2)
```

```python
print(a2.dtype)
```

### empty function

```python
x9 = np.empty([10,3], dtype=int)
print(x9)
```

```python
z1 = np.empty([1,10], dtype=int)
print(z1)
```

### zeros function

```python
z3 = np.zeros(5, dtype=int)
print(z3)
```

```python
z2 = np.zeros([10,2,3], dtype=int)
print(z2)
```

```python
z4 = np.zeros([2,2], dtype=np.int_)
print(z4)
```

### ones function

```python
np.ones(5, dtype=int)
```

```python
x10 = np.ones([2,10,10], dtype=int)
print(x10)
```

```python
x11 = np.ones([2,10,10], dtype=np.float64)
print(x11)
```

### arange function

```python
b1 = np.arange(10,20,2)
print(b1)
```

```python
b2 = np.arange(10,20,0.5)
print(b2)
```

### linspace function

```python
c1 = np.linspace(start=2.0, stop=4.0, num=50)
print(c1)
```

```python
c2 = np.linspace(start=2.0, stop=4.0, num=5, endpoint=False)
print(c2)
```
